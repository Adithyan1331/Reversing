13. Check Prime Number
Determine whether a given number is prime.

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime Number")
else:
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not a Prime Number")
```

14. Next Prime Number
Given an integer N, find and print the next prime number greater than N

```python
n = int(input("Enter a number: "))

num = n + 1

while True:
    is_prime = True

    if num <= 1:
        is_prime = False
    else:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break

    if is_prime:
        print("Next prime number:", num)
        break

    num += 1
```

18. LCM of Two Numbers
Find the Least Common Multiple (LCM) of two numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

x, y = a, b

while y != 0:
    x, y = y, x % y

gcd = x

lcm = (a * b) // gcd

print("LCM =", lcm)
```

19. Count Digits
Count the total number of digits in a given integer.

```python
num = int(input("Enter a number: "))

num = abs(num)  # handle negative numbers
count = 0

if num == 0:
    count = 1
else:
    while num > 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```

20. Reverse a Number
Reverse the digits of a given integer.

```python
num = int(input("Enter a number: "))

rev = 0
n = abs(num)

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10

# handle negative numbers
if num < 0:
    rev = -rev

print("Reversed number:", rev)
```

27. Armstrong Number
Check whether a given number is an Armstrong number

```python
num = int(input("Enter a number: "))

# Number of digits
digits = len(str(num))

# Calculate sum of each digit raised to the power of digits
sum_of_powers = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** digits
    temp //= 10

# Check Armstrong condition
if num == sum_of_powers:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

28. Palindrome Number
Check whether a number remains the same when its digits are reversed.

```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print(original, "is a palindrome number")
else:
    print(original, "is not a palindrome number")
```

29. Strong Number
Check whether a number is a Strong Number (sum of factorials of digits equals the number)

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter a number: "))

temp = num
sum_fact = 0

while temp > 0:
    digit = temp % 10
    sum_fact += factorial(digit)
    temp //= 10
if sum_fact == num:
    print(num, "is a Strong Number")
else:
    print(num, "is not a Strong Number")
```

34. Decimal to Binary Converter
Convert a decimal number into binary without using built-in conversion functions

```python
num = int(input("Enter a decimal number: "))

if num == 0:
    print("Binary: 0")
else:
    binary = ""

    while num > 0:
        remainder = num % 2
        binary = str(remainder) + binary
        num //= 2

    print("Binary:", binary)
```

40. Mini ATM Simulation
Develop a menu-driven ATM simulation supporting balance inquiry, deposit, withdrawal, and exit

```python
balance = 1000

while True:
    print("\n===== MINI ATM MENU =====")
    print("1. Balance Inquiry")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Current Balance:", balance)

    elif choice == 2:
        amount = float(input("Enter deposit amount: "))
        if amount > 0:
            balance += amount
            print("Amount deposited successfully!")
        else:
            print("Invalid amount")

    elif choice == 3:
        amount = float(input("Enter withdrawal amount: "))
        if amount <= balance and amount > 0:
            balance -= amount
            print("Please collect your cash")
        else:
            print("Insufficient balance or invalid amount")

    elif choice == 4:
        print("Thank you for using the ATM. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
